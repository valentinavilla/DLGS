```python
import gym
import numpy as np

env = gym.make("CartPole-v1")
state_space_dim = env.observation_space.shape[0]
action_space_dim = env.action_space.n
```

```python
class EchoStateNetwork:
    def __init__(self, input_size, reservoir_size, output_size, connectivity=0.1, spectral_radius=0.9):
        self.input_size = input_size
        self.reservoir_size = reservoir_size
        self.output_size = output_size

        # Initialize the reservoir with random values
        self.reservoir = np.random.rand(reservoir_size, reservoir_size) - 0.5
        self.reservoir *= connectivity

        # Compute the spectral radius of the reservoir
        max_eigenvalue = max(abs(np.linalg.eigvals(self.reservoir)))
        self.reservoir /= max_eigenvalue / spectral_radius

        # Initialize the output weights
        self.W_out = np.random.rand(output_size, input_size + reservoir_size + 1) - 0.5
```

```python
def run(self, input_data, initial_state, time_steps):
        reservoir_states = np.zeros((time_steps, self.reservoir_size))
        output = np.zeros((time_steps, self.output_size))

        x = np.zeros(self.reservoir_size)

        for t in range(time_steps):
            u = np.concatenate((input_data[t], x, [1]))
            x = np.tanh(np.dot(self.reservoir, x) + np.dot(self.W_out, u))
            reservoir_states[t] = x
            output[t] = np.dot(self.W_out, u)

        return output, reservoir_states
```

```python
# Define the input, reservoir, and output sizes
input_size = 4
reservoir_size = 100
output_size = 2

# Create an instance of the EchoStateNetwork
esn = EchoStateNetwork(input_size, reservoir_size, output_size)

# Generate some example input data (replace with your data)
time_steps = 300
input_data = np.random.rand(time_steps, input_size)

# Provide an initial state for the reservoir
initial_state = np.random.rand(reservoir_size)

# Run the ESN on the input data
output, reservoir_states = esn.run(input_data, initial_state, time_steps)
```
